```python
# average value by proffesssion 
average_salary_by_profession = df.groupby('job_title')['salary_in_usd'].agg( ['mean', 'min', 'max']).round().rename(columns={'mean': 'mean_salary', 'min': 'min_salary', 'max': 'max_salary'})

sorted_average_salary_by_profession = average_salary_by_profession.sort_values(by='mean_salary', ascending=False)

sorted_average_salary_by_profession
```

```python
#salaries for each job according to every year
average_salary_by_year = df.groupby(['work_year', 'job_title'])['salary_in_usd'].agg( ['mean', 'min', 'max']).round().rename(columns={'mean': 'mean_salary', 'min': 'min_salary', 'max': 'max_salary'})
average_salary_by_year
```

```python
# the dict of jobs with their numbers of appearing
job_title_counts = df['job_title'].value_counts()

job_title_dict = job_title_counts.to_dict()
job_title_dict
```

```python
#count the most popular professions
most_3_popular = {}
c = 0
for k, v in job_title_dict.items():
    most_3_popular[k] = v
    c += 1
    if c == 3:
        break
most_3_popular
    
```

```python
# prof. levels with their numbers of appearances
level_title_counts = df['experience_level'].value_counts()

level_title_dict = level_title_counts.to_dict()
level_title_dict
```

```python
# jobs with the appearances of less than 11
def get_lessneed_jobs(d):
    l = []
    for k, v in d.items():
        if v <= 10:
            l.append(k)
    return l

# jobs with the number of appearances between top3 and lowest
def count_avr_jobs(d):
    l = []
    for k, v in d.items():
        if v > 10 and v < 2189:
            l.append(k)
    return l
```

```python
# results of 2 functions above
most_unclaimed_jobs = get_lessneed_jobs(job_title_dict)
avr_jobs = count_avr_jobs(job_title_dict)
num_avr_jobs = len(avr_jobs)
```

```python
print(avr_jobs)
print(num_avr_jobs)
print(len(most_unclaimed_jobs))
```

```python
# nested dictionary in format { year : { job : number of a job in this year } }
def nested_dict_counts(df, year_column, job_column):
    nested_dict = {}
    
    years = df[year_column].unique()

    for year in years:
        year_df = df[df[year_column] == year]
        job_counts = year_df[job_column].value_counts().to_dict()
        nested_dict[year] = job_counts
    
    return nested_dict

year_job_dict = nested_dict_counts(df, 'work_year', 'job_title')
```

```python
# excluding values for Data Engineer
def get_top_engineers(d):
    data_engineer_counts = {}

    for year, job_counts in d.items():
        data_engineer_count = job_counts.get('Data Engineer', 0)
        data_engineer_counts[year] = data_engineer_count

    return data_engineer_counts
```

```python
# excluding values for Data Analyst
def get_top_analytiscs(d):
    data_analyst_counts = {}

    for year, job_counts in d.items():
        data_analyst_count = job_counts.get('Data Analyst', 0)
        data_analyst_counts[year] = data_analyst_count

    return data_analyst_counts
```

```python
# excluding values for Data Scientist
def get_top_datascience(d):
    data_scientist_counts = {}

    for year, job_counts in d.items():
        data_scientist_count = job_counts.get('Data Scientist', 0)

        data_scientist_counts[year] = data_scientist_count

    return data_scientist_counts
```

```python
# saving the results
engi_statistics = dict(sorted(get_top_engineers(year_job_dict).items()))
analytic_statistics = dict(sorted(get_top_analytiscs(year_job_dict).items()))
dsience_statistics = dict(sorted(get_top_datascience(year_job_dict).items()))
```
